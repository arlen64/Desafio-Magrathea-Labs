```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
from sqlalchemy.exc import OperationalError
```

### Carregar o arquivo 'NCM' no banco de dados

```python
# Caminho do csv
caminho_NCM = 'NCM.csv'

# Retornar se o carregamento ocorreu com sucesso
try:
    df_table = pd.read_csv(caminho_NCM, encoding='latin1', sep=';')
    print("Arquivo CSV carregado com sucesso!")
except Exception as e:
    print("Ocorreu um erro ao carregar o arquivo CSV:", str(e))

# Conectar ao banco de dados PostgreSQL

    # Substitua 'usuario', 'senha', 'host', 'porta' e 'nome_do_banco' pelos seus próprios valores

engine_ncm = create_engine('postgresql://postgres:1547@localhost:5432/dados_gov')

try:
    # Tente conectar
    conn = engine_ncm.connect()
    conn.close()
    print("Conexão bem-sucedida!")
except OperationalError as e:
    print(f"Falha na conexão: {e}")

# Salvar o DataFrame como uma tabela no banco de dados PostgreSQL

    # Substitua 'nome_da_tabela' pelo nome da tabela que você deseja criar
try:
    df_table.to_sql('table_names', engine_ncm, if_exists='replace', index=False)
    print("Dados armazenados no banco de dados com sucesso!")
except Exception as e:
    print("Ocorreu um erro ao armazenar os dados no banco de dados:", str(e))
```

## Carregar o arquivo 'EXP_COMPLETA' no banco de dados

#### Carregando arquivos em Lotes.

```python
# Caminho do csv
caminho_exp= 'EXP_COMPLETA.csv'

# Conectar ao banco de dados PostgreSQL usando SQLAlchemy
engine = create_engine('postgresql://postgres:1547@localhost:5432/dados_gov')

# Definir o nome da tabela no banco de dados onde os dados serão armazenados
nome_tabela = 'exportacoes'

# Determinar o tamanho do lote para carregar os dados
chunksize = 50000  # Ajuste conforme necessário

# Carregar os dados do DataFrame para o banco de dados em lotes menores
try:
    # Loop através do DataFrame em lotes e carregar cada lote para o banco de dados
    for i, chunk in enumerate(pd.read_csv(caminho_exp, encoding='latin1', sep=';', chunksize=chunksize)):
        chunk.to_sql(nome_tabela, engine, if_exists='replace', index=False)
        print(f"Lote {i+1} carregado com sucesso.")
    
    print("Todos os dados foram carregados com sucesso para o PostgreSQL!")
except Exception as e:
    print("Ocorreu um erro ao carregar os dados no banco de dados:", str(e))

# O carregamento completo durou 21 minutos.
```
